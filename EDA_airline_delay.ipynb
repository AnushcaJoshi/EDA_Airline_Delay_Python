**Carrier and number of delay**

```python
x = "carrier"
y = "carrier_delay"

plt.figure(figsize=(15, 6))

sns.barplot(x =x, y =y, data = df, palette = "pastel", hue ="year")

plt.title("Carrier delay")
plt.xlabel("Carrier Name")
plt.ylabel("Carrier Delay")

plt.show()
```

```python
sns.scatterplot(x = 'carrier', data = df,y = 'arr_delay', palette = "pastel",hue = 'year')

plt.xlabel("carrier")
plt.ylabel("Delay Time (minutes)")
```
